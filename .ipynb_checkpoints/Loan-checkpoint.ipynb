```python
# import libraries
import pandas as pd
import numpy as np

X = dfLoan[['ApplicantIncome', 'Loan_Amount_Term']].to_numpy()

# Use Scikit-Learn
from sklearn.cluster import KMeans
kmean = KMeans(n_clusters = 3)
kmean.fit(X)
```

```python
kmean.cluster_centers_
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
centroid = kmean.cluster_centers_
plt.scatter(centroid[:, 0], centroid[:, 1], s=50, c='b')
plt.show()
```
